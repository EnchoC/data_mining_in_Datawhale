```python
def get_data_msg(data, sort_str='null'):
    data_des = data.describe(percentiles=[], include='all').T
    data_des['null'] = data.shape[0] - data_des['count']
    data_des['miss_%'] = data_des['null'] / data.shape[0]
    data_des['dtype'] = data.dtypes
    return data_des.sort_values(sort_str, ascending=False)
get_data_msg(data).head()
```

- unique<10 的用众数填充
- unique>10 的用平均数填充

```python
def parse_na1(tmp):
    d = {}
    for i in tmp.columns:
        if tmp[i].nunique() < 10:
            status1 = tmp[data['status']==1][i].value_counts().index[0]
            status0 = tmp[data['status']==0][i].value_counts().index[0]
        elif tmp[i].nunique() > 10:
            status1 = int(tmp[data['status']==1][i].dropna().mean())
            status0 = int(tmp[data['status']==0][i].dropna().mean())
        d[i] = {'1':status1, '0':status0}
    return d
```

```python
na1_values = parse_na1(na1)
```

```python
for i in na1.columns:
    data.loc[(data[i].isnull()) & (data['status']==0), i] = na1_values[i]['0']
    data.loc[(data[i].isnull()) & (data['status']==1), i] = na1_values[i]['1']
```

- 2%~10%的处理

用随机森林预测填充这部分数据

```python
na10 = data[data.columns[(data.isnull().sum(0)/data.shape[0] <= 0.1)
            & (data.isnull().sum(0)/data.shape[0] >= 0.01)]]
```

```python
from datetime import datetime
import time
def parse_time(ts):
    try:
        dt = time.mktime(datetime.strptime(ts, "%Y-%m-%d").timetuple())
    except:
        dt = np.nan
    return dt
```

```python
for col in ['latest_query_time','loans_latest_time']:
    na10[col] = na10[col].apply(parse_time)
```

```python
from sklearn.ensemble import RandomForestRegressor
def predict_nan(col):
    nan_exist = data.loc[data[col].notnull()].drop(na10.columns, axis=1)
    y_nan = na10[col][na10[col].notnull()]
    nan_null = data.loc[data[col].isnull()].drop(na10.columns, axis=1)
    x = nan_exist.values
    y = y_nan.values
    rfr_model = RandomForestRegressor(n_estimators=100)
    rfr_model.fit(x, y)
    col_hat = rfr_model.predict(nan_null.values)
    col_hat = col_hat.astype(int)
    na10.loc[(na10[col].isnull()), col] = col_hat
    print('{:-^40}'.format('column[%s] done'%col))
```

```python
[predict_nan(c) for c in na10.columns]
```

```python
for col in ['latest_query_time','loans_latest_time']:
    na10[col] = pd.to_datetime(na10[col], unit='s').dt.strftime('%Y-%m-%d')
```

```python
for col in na10.columns:
    data[col] = na10[col]
```
